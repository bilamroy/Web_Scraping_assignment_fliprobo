#Answer 1

```python
header = []
def scrap1(url):
    req = requests.get(url)
    content = req.content
    soup = BeautifulSoup(content)
    headings=soup.find('tr').find_all("h2")
    for head in headings:
        text = head.find('span',class_='mw-headline')
        header.append(text.get_text())
    return (header)
```

```python
scrap1("https://en.wikipedia.org/wiki/Main_Page")
```

#Answer 2

```python
releasing_year=[]
movie_title=[]
imdb_rating=[]

def scrap2(url):
    req = requests.get(url)
    content = req.content
    soup = BeautifulSoup(content)
    
    year = soup.select("td.titleColumn")
    released_in = year[0:10]

    movie_name = soup.select("td.titleColumn a")
    movie_tag=movie_name[0:10]

    ratings = soup.select("td.posterColumn span[name=ir]")
    rating_tag = ratings[0:10]

    for year in released_in:
            year_tag = year.get_text()
            split = year_tag.split()
            year = split[-1]
            releasing_year.append(year)

    for name in movie_tag:
            movie_name_title = name.get_text()
            movie_title.append(movie_name_title)

    for rating in rating_tag:
            movie_rating = float((rating['data-value']))
            rate='%.1f'%movie_rating
            imdb_rating.append(rate)

    movies = pd.DataFrame({'Movie':movie_title,
                               'Year':releasing_year,
                               'IMDB_Rating':imdb_rating,})
    
    movies.to_csv('movies.csv')
    
    return (movies)
```

```python
scrap2("https://www.imdb.com/chart/top")
```

#Answer 4

```python
book_name=[]
author_name=[]
genre=[]
book_review=[]

def scrap4(url):
    req = requests.get(url)
    content = req.content
    soup = BeautifulSoup(content)

    booknames = soup.select("div.flex-article-content h4")
    booknames = booknames[0:5]

    author = soup.select("div.flex-article-content p.sans")
    author = author[0:5]

    genres = soup.select("div.flex-article-content p.genre-links")
    genres = genres[0:5]

    review = soup.select("div.read-full")
    review = review[0:5]

    for bk_name in booknames:
            book_name_title = bk_name.get_text().replace("\n", " ")
            book_name.append(book_name_title)

    for name in author:
            bk_author = name.get_text().replace("\n", " ")
            author_name.append(bk_author)

    for gn_name in genres:
            bk_genre = gn_name.get_text().replace("\n", " ")
            genre.append(bk_genre)

    for link in review:
        book_review.append(link.a['href'])

    book = pd.DataFrame({'Name':book_name,
                         'Author':author_name,
                         'Genre':genre,
                         'Full Book Review links(https://bookpage.com/)':book_review,})

    return (book)
```

```python
scrap4("https://bookpage.com/reviews")
```

#Answer 5(i)

```python
def scrap5_i(url):
    req = requests.get(url)
    content = req.content
    soup = BeautifulSoup(content)
    
    
    teams = []
    top_10 = soup.select("tbody span.u-hide-phablet")
    top_10 = top_10[0:10]

    for top_teams in top_10:
            top_10 = top_teams.get_text()
            teams.append(top_10)


    no_of_matches = []    
    matches = soup.select("tbody td.rankings-block__banner--matches")
    matches = matches[0].get_text()
    no_of_matches.append(matches)

    matches = soup.select("tbody td.table-body__cell.u-center-text")
    matches = matches[0:18:2]

    for counts in matches:
            matches = counts.get_text()
            no_of_matches.append(matches)


    total_points = []    
    points = soup.select("tbody td.rankings-block__banner--points")
    points = points[0].get_text()
    total_points.append(points)

    points = soup.select("tbody td.table-body__cell.u-center-text")
    points = points[1:19:2]

    for counts in points:
            points = counts.get_text()
            total_points.append(points)


    total_ratings = []    
    ratings = soup.select("tbody td.rankings-block__banner--rating.u-text-right")
    #print(ratings)
    ratings = ratings[0].get_text().replace("\n                            ",'').replace("\n",'')
    total_ratings.append(ratings)

    ratings = soup.select("tbody td.table-body__cell.u-text-right.rating")
    ratings = ratings[0:9]

    for counts in ratings:
            ratings = counts.get_text()
            total_ratings.append(ratings)
    
    
    odi_top_10 = pd.DataFrame({'Team':teams,
                                'Matches':no_of_matches,
                                'Points':total_points,
                                'Ratings':total_ratings})

    return (odi_top_10)
```

#Answer 7

```python
Name = []
Price = []
Images = []
Ratings = []


def scrap7(url):
    req = requests.get(url)
    content = req.content
    soup = BeautifulSoup(content)

    model_name = soup.select("h2 a")
    model_name = model_name[0:10]

    for model in model_name:
            model_names = model.get_text().replace("\n", "")
            Name.append(model_names)
            
            
    model_prices = soup.select("div.a-row span.a-price-whole")
    model_prices = model_prices[0:10]

    for model in model_prices:
            mdl_price = model.get_text()
            Price.append(mdl_price)
            
            
    model_image = (soup.find_all('img'))
    model_image = model_image[0:10:3]
    #for model in model_image:
    #        model_image = model.get_text()
    Images.append(model_image)
    
    
    model_rating = soup.select("i span.a-icon-alt")
    model_rating = model_rating[4:14]

    for model in model_rating:
            rate = model.get_text()
            Ratings.append(rate)
            
            
    model_specifications = pd.DataFrame.from_dict({'Name':Name,
                                         'Price':Price,
                                         'Image Links':Images,
                                         'Ratings':Ratings}, orient ='index')
    model_specifications = model_specifications.transpose()

    return (model_specifications)
```

```python
scrap7("https://www.amazon.in/s?k=mobile+phones+under+20000&ref=nb_sb_noss_1")
```
