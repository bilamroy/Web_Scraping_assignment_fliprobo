#Answer 1

```python
header = []
def scrap1(url):
    req = requests.get(url)
    content = req.content
    soup = BeautifulSoup(content)
    headings=soup.find('tr').find_all("h2")
    for head in headings:
        text = head.find('span',class_='mw-headline')
        header.append(text.get_text())
    return (header)
```

```python
scrap1("https://en.wikipedia.org/wiki/Main_Page")
```

#Answer 2

```python
releasing_year=[]
movie_title=[]
imdb_rating=[]

def scrap2(url):
    req = requests.get(url)
    content = req.content
    soup = BeautifulSoup(content)
    
    year = soup.select("td.titleColumn")
    released_in = year[0:10]

    movie_name = soup.select("td.titleColumn a")
    movie_tag=movie_name[0:10]

    ratings = soup.select("td.posterColumn span[name=ir]")
    rating_tag = ratings[0:10]

    for year in released_in:
            year_tag = year.get_text()
            split = year_tag.split()
            year = split[-1]
            releasing_year.append(year)

    for name in movie_tag:
            movie_name_title = name.get_text()
            movie_title.append(movie_name_title)

    for rating in rating_tag:
            movie_rating = float((rating['data-value']))
            rate='%.1f'%movie_rating
            imdb_rating.append(rate)

    movies = pd.DataFrame({'Movie':movie_title,
                               'Year':releasing_year,
                               'IMDB_Rating':imdb_rating,})
    
    movies.to_csv('movies.csv')
    
    return (movies)
```

```python
scrap2("https://www.imdb.com/chart/top")
```

#Answer 4

```python
book_name=[]
author_name=[]
genre=[]
book_review=[]

def scrap4(url):
    req = requests.get(url)
    content = req.content
    soup = BeautifulSoup(content)

    booknames = soup.select("div.flex-article-content h4")
    booknames = booknames[0:5]

    author = soup.select("div.flex-article-content p.sans")
    author = author[0:5]

    genres = soup.select("div.flex-article-content p.genre-links")
    genres = genres[0:5]

    review = soup.select("div.read-full")
    review = review[0:5]

    for bk_name in booknames:
            book_name_title = bk_name.get_text().replace("\n", " ")
            book_name.append(book_name_title)

    for name in author:
            bk_author = name.get_text().replace("\n", " ")
            author_name.append(bk_author)

    for gn_name in genres:
            bk_genre = gn_name.get_text().replace("\n", " ")
            genre.append(bk_genre)

    for link in review:
        book_review.append(link.a['href'])

    book = pd.DataFrame({'Name':book_name,
                         'Author':author_name,
                         'Genre':genre,
                         'Full Book Review links(https://bookpage.com/)':book_review,})

    return (book)
```

```python
scrap4("https://bookpage.com/reviews")
```
